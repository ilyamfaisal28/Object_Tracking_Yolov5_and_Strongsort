# Making Graphic from CSV file

```python
import matplotlib.ticker as ticker
import pandas as pd
import matplotlib.pyplot as plt


# df = pd.read_csv('data/timestamp01.csv', parse_dates=True, index_col='TIMESTAMP', sep=";")
df = pd.read_csv('/content/Exported_Data.csv', parse_dates=True, index_col='Timestamp', sep=",")
df
```

```python
f, ax = plt.subplots(1,1)

# ax.plot(df.resample('30s').size())
ax.plot(df.resample('30s').size().between_time('10:15:0', '10:16:18'))

plt.setp(ax.xaxis.get_majorticklabels(), rotation=30, ha="right")
plt.show()
```
